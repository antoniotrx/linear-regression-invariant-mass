```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/final_data.csv")
yPredictions = pd.read_csv("../data/yPredictions.csv").to_numpy()
y_test = pd.read_csv("../data/y_test.csv").to_numpy()
```

```python
yPredictions = yPredictions.reshape(1, -1)[0]
y_test = y_test.reshape(1, -1)[0]
```

```python
# Scatterplot

plt.figure(figsize=(8, 6))
sns.scatterplot(x=yPredictions, y=y_test)
plt.xlabel("M - Predictions")
plt.ylabel("M - Actual values")
plt.title("Predicted VS Actual values")
```

```python
# Residual plot
plt.figure(figsize=(8,6))
sns.residplot(x=yPredictions, y=y_test)
plt.xlabel("M - Predictions")
plt.ylabel("Residuals")
plt.title("Residual plot")
```

On this plot, on the x-axis are the predicted values for invaraint mass (M) and on the y-axis are the residuals, the difference between the predicted and the actual value for M. We can see that on different ranges of M we have different residuals, while ideally we should have randomlly scattered points arounf the 0 dashed line. Also we can notice a curved shape which may be result of that the invariant mass doesn't depend linearly but its true dependence is nonlinear and given by $M = \sqrt{(E_1 + E_2)^2 - (\vec{p_1} + \vec{p_2})^2}$. So nonlinear models like polynomial regression would likely perform better on this problem.
