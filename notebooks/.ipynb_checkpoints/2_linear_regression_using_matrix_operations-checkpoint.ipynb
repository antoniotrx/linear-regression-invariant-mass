# Linear Regression Model using Matrix Operations

```python
# split the predictors and the target variable in the data
y = data["M"]
x = data.drop(columns=["M", "Run", "Event"], axis=1)
```

```python
# train-test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
x_train.shape
```

```python
# Prepend 1 to each row of the predictors array
ones_column_train = np.ones((x_train.shape[0], 1))
ones_column_test = np.ones((x_test.shape[0], 1))
x_train = np.hstack([ones_column_train, x_train])
x_test = np.hstack([ones_column_test, x_test])
```

```python
x_train.shape, x_test.shape
```

```python
# Transpose x which is a 99892x21 matrix into a 21x99892 matrix
xTransposed = np.transpose(x_train)
```

```python
# Calculate the beta column vector
beta = np.linalg.pinv(xTransposed.dot(x_train)).dot(xTransposed).dot(y_train)
```

```python
beta
```

```python
# Calculate predictions for the target variable
yPredictions = x_test.dot(beta)
```

```python
yPredictions[:5]
```

```python
# Real values of target variable
y_test[:5]
```

```python
ae_series = np.abs(yPredictions - y_test)
se_series = np.square(yPredictions - y_test)
mae = ae_series.mean()
mse = se_series.mean()
rmse = np.sqrt(mse)
```

```python
print(f"Mean squared error: {mse};\nRoot Mean Squared Error: {rmse};\nMean Absolute Error: {mae}.")
```
